```python
from geopy.geocoders import Nominatim 
```

```python
df = pd.read_csv('/content/drive/MyDrive/시소프로젝트/NEW(공모전용)/전처리용/연면적.csv')
df.head(1)
```

```python
geolocoder = Nominatim(user_agent = 'South Korea', timeout = None)

def geocoding(address) : 
    try : 
        geo= geolocoder.geocode(address)
        crd = (str(geo.latitude), str(geo.longitude))
        return crd
    
    except : 
        return (0,0)
```

```python
lats = []
logs = []

for address in df['주소'] : 
    crd = geocoding(address)
    lats.append(crd[0])
    logs.append(crd[1])

lats = pd.Series(lats)
logs = pd.Series(logs)
```

```python
df = pd.concat([df, lats], axis = 1)
df = pd.concat([df, logs], axis = 1)
df.head(5)
```

```python
df.columns = ['자치구', '동', '시설명', '주소', '연면적', '위도', '경도']
```
