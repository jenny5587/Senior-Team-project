## 주변특성 pca

```python
# StandardScaler와 logScale을 활용하여 연속형 변수 정규화를 진행한다.
from sklearn.preprocessing import StandardScaler, MinMaxScaler
number = df1[['공원수','도서관수','배움터수','병원수','대중교통']]

scaled_data = StandardScaler().fit_transform(np.log1p(number))
data_scale_log=pd.DataFrame(data=scaled_data, columns=number.columns)
```

```python
scale_df = data_scale_log
scale_df[:5]
```

```python
from sklearn.decomposition import PCA
# 주변특성 pca (설명력 80%)
pca = PCA(n_components=0.8)
pca.fit(number)
pca_df1 = pd.DataFrame((pca.transform(number)))
```

```python
pca_df1.shape
```

```python
pca_result = pd.DataFrame(data = pca.components_, columns = number.columns)
pca_result.transpose()
```
